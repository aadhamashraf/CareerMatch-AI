```python
import openai
import gradio as gr
import os



os.environ["GROQ_API_KEY"] = "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"

client = openai.OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=os.environ["GROQ_API_KEY"]
)
MODEL = "llama-3.1-8b-instant"


def polish(text, job="Data Scientist"):
    prompt = f"""You are the world's #1 resume writer for {job} roles at Google, Meta, and top startups.

Rewrite this bullet point to be:
• One line only
• Start with the strongest action verb possible
• Quantify impact with realistic impressive numbers
• 100% ATS-friendly and recruiter-approved
• Remove every cliché

Original:
{text}

Improved bullet only (no quotes, no explanation):"""

    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3,
        max_tokens=120
    )
    return response.choices[0].message.content.strip()
```

```python
def ui(original, job):
    if not original.strip():
        return "error"
    return polish(original, job)

iface = gr.Interface(
    fn=ui,
    inputs=[
        gr.Textbox(lines=3, placeholder="مثال: Worked on some Python projects...", label="الجملة الأصلية"),
        gr.Textbox(value="Data Scientist", label="الوظيفة المستهدفة")
    ],
    outputs=gr.Textbox(label="the strong version "),
    title="Resume Polisher Pro 2025",
    examples=[
        ["Helped develop machine learning models", "Machine Learning Engineer"],
        ["Did some data analysis using Excel", "Data Analyst"],
        ["Managed social media pages", "Growth Marketing"]
    ]
)

iface.launch(share=True)
```
